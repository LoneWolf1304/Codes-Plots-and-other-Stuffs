```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','notebook','grid'])
```

```python
x=np.arange(-10,10,0.01)
```

```python
#Setting parameters to arbitrary value 
m=1
w=1
h=1
```

```python
#Defining the Ground State of the Harmonic oscillator
GS=((m*w/(np.pi*h))**(1/4))*np.exp(-m*w*(x**2)/(2*h))
```

```python
#Defining a function which numerically calculates the derivative
def derivative1(f):
    df=f[0]
    dx=0.01
    f_d=np.zeros((2000,))
    for i in range(1999):
        f_d[i]=((f[i+1]-f[i])/dx)
    return f_d
```

```python
#Calculating the higher states by applying the creation operator repetitively 
eig=GS
for i in range(5):
    plt.plot(x,eig+5*i)
    eig_der=derivative1(eig)
    eig=(x*eig-eig_der)/(np.sqrt(2*m*h*w))
plt.plot(x,x**2, lw=0.8 )
plt.legend(['n=0', 'n=1','n=2','n=3','n=4'],fontsize="10", loc='upper right', title='Eigenstates',title_fontsize=10)
plt.ylim((0,25))
plt.ylabel(r'$\psi(x)$')
plt.yticks([])
```
